```python
import pandas as pd
```

```python
# row count check
def get_row_counts(full_table_name):
    df = spark.read.table(full_table_name)
    return df.count()

```

```python
# tables to  scan - table_name: count_expected
tables_to_scan = {
    "data_sentinals.bronze.raw_customers": 99441,
    "data_sentinals.bronze.raw_geolocation": 1000163,
    "data_sentinals.bronze.raw_olist_sellers": 3095,
    "data_sentinals.bronze.raw_orders": 99441,
    "data_sentinals.bronze.raw_order_items": 112650,
    "data_sentinals.bronze.raw_order_payments": 103886,
    "data_sentinals.bronze.raw_order_reviews": 99224,
    "data_sentinals.bronze.raw_products": 32951,
    "data_sentinals.bronze.raw_product_names_translation": 71
}

```

```python
# empty list to hold our audit results
audit_results = []

for item in tables_to_scan:
    
    expected_count = int(tables_to_scan[item])
    actual_count = int(get_row_counts(item))
    
    difference = actual_count - expected_count
    status = "✅ PASS" if difference == 0 else "❌ FAIL"
    
    # 4. Appending the record to our report
    audit_results.append({
        "Table Name": item,
        "Status": status,
        "Expected Count": expected_count,
        "Actual Count": actual_count,
        "Difference": difference
    })

audit_df = spark.createDataFrame(audit_results)

# 6. Display the rich table in Databricks
display(audit_df)
```
